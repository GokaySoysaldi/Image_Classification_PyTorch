##  Transform images to tensors and resize them if needed

```python
transform = transforms.Compose(
    [transforms.ToTensor(),  # Transfer images to tensors.
     torchvision.transforms.Resize((32, 32))]) # Images are 32*32. Height = 32, Width = 32.
```

## Downloading and splitting CIFAR10 dataset as train and testset

```python
batch_size = 128

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
```

## Creating validation set

```python
train_size = int(0.8 * 50000)
train_size
```

```python
valid_size = len(trainset) - train_size
train_set, validset = torch.utils.data.random_split(trainset, [train_size, valid_size]) # Randomly splits trainset.
```

```python
# Test set.
test_classes = [label for _, label in testset]
Counter(test_classes)
```

```python
# Validation set.
val_classes = [label for _, label in validset]
Counter(val_classes)
```

## Creating Dataloaders

```python
trainloader = torch.utils.data.DataLoader(train_set, batch_size=batch_size,
                                          shuffle=True, num_workers=0)

testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=True, num_workers=0)

validloader = torch.utils.data.DataLoader(validset, batch_size=batch_size,
                                         shuffle=True, num_workers=0)
```

```python
# Define labels for model.
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

# Convolutional Neural Network

## Example CNN Architecture Used

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.cnn_layers = nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=32, kernel_size=(3, 3), padding=5),   # 32*32
        nn.ReLU(),
        nn.Conv2d(in_channels=32, out_channels=64, kernel_size=(3, 3), padding=0),  # 32*32
        nn.Dropout(0.2),
        nn.ReLU(),
        nn.BatchNorm2d(64), # Batch Normalization 2d cause of 4d input.
        nn.MaxPool2d(kernel_size=(2, 2)), #16*16
        
            
        nn.Conv2d(in_channels=64, out_channels=128, kernel_size=(3, 3), padding=0), # 16*16
        nn.ReLU(),
        nn.Conv2d(in_channels=128, out_channels=128, kernel_size=(3,3), padding=1), # 16*16
        nn.Dropout(0.2),
        nn.ReLU(),
        nn.BatchNorm2d(128),
        nn.MaxPool2d(kernel_size=(2, 2)), # 8*8
            
            
        nn.Conv2d(in_channels=128, out_channels=256, kernel_size=(3,3), padding=0), # 8*8
        nn.ReLU(),
        nn.Conv2d(in_channels=256, out_channels=256, kernel_size=(3,3), padding=1), # 8*8
        nn.Dropout(0.2),
        nn.ReLU(),
        nn.BatchNorm2d(256),
        nn.MaxPool2d(kernel_size=(2, 2)), #4*4
        ).to(device)
        
        self.classifier = nn.Sequential(
        nn.Flatten(),
        nn.Dropout(0.2),
        nn.Linear(2304, 256),  # 256*4*4 = 4096.
        nn.ReLU(),
            
        nn.BatchNorm1d(256),  # Batch Norm 1d needed after flatten.  
            
        nn.Linear(256, 10),   # 10 output layers cause we have 10 labels.
        #nn.Softmax(),        # Dont need Softmax cause we use CrossEntropLoss as criterion.
        ).to(device)
        
    def forward(self, x):
        x = self.cnn_layers(x)
        x = self.classifier(x)
        return x
    
net = CNN().to(device)
```

```python
#Summary of model.
summary(model=net, input_size=(3, 32, 32), batch_size=128)
```

## Clears cache if needed

```python
import gc
import torch
del net
gc.collect()
torch.cuda.empty_cache()
```

```python
criterion = nn.CrossEntropyLoss()

lear_rate = 0.001 # Learning Rate

optimizer = optim.Adam(net.parameters(), lr = lear_rate, weight_decay=0.04) #L2 Regularization weight = 0.04

epoch = 10

# Learning Rate Scheduler.
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer,factor=0.1,patience=1,verbose=True)
```

## Training

```python
training_loss = []
val_loss = []
for epc in range(0,epoch):  # Loop over the dataset multiple times

    running_trainloss = []
    running_valloss = []
    #net.train()
    with torch.enable_grad():
        for i, data in enumerate(tqdm(trainloader, 0)):
            # Get the inputs; data is a list of [inputs, labels]
            inputs, labels = data
            inputs = inputs.to(device)
            labels = labels.to(device)
            #inputs, labels = data[0].to(device), data[1].to(device)
            #inputs, labels = inputs.to(device), labels.to(device)
            # Zero the parameter gradients
            optimizer.zero_grad()

            # Forward + backward + optimize
            outputs = net(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            
            #running_loss += loss.item()
            running_trainloss.append(loss.item())

    #net.eval()     # Optional when not using Model Specific layer
    with torch.no_grad():
        for data, labels in validloader:
            if torch.cuda.is_available():
                data, labels = data.to(device), labels.to(device)

            target = net(data)
            loss = criterion(target,labels)
            running_valloss.append(loss.item())
    
    trainloss = sum(running_trainloss) / len(running_trainloss)
    
    scheduler.step(trainloss)
    
    valloss = sum(running_valloss) / len(running_valloss)
    # Print statistics
    print('[%d, %5d] train loss: %.3f val loss: %.3f' %
          (epc + 1, i + 1, trainloss, valloss))
    #running_loss = 0.0
    training_loss.append(trainloss)
    val_loss.append(valloss)
    
    
print('Finished Training')
```

## Loss Curves

```python
plt.title("Training and Validation Loss Curves")
plt.plot(val_loss,label="val")
plt.plot(training_loss,label="train")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

## Confusion Matrix

```python
y_pred = []
y_true = []

# Iterate over test data
for inputs, labels in testloader:
    inputs = inputs.to(device)
    output = net(inputs) # Feed Network
        

    output = (torch.max(torch.exp(output), 1)[1]).data.cpu().numpy()
    y_pred.extend(output) # Save Prediction
        
    labels = labels.data.cpu().numpy()
    y_true.extend(labels) # Save Truth

# Build confusion matrix
cf_matrix = confusion_matrix(y_true, y_pred)
df_cm = pd.DataFrame(cf_matrix/np.sum(cf_matrix) *10, index = [i for i in classes],
                     columns = [i for i in classes])

#df_cm = pd.DataFrame(cf_matrix/np.sum(cf_matrix) *10, index = [i for i in classes],
                     #columns = [i for i in classes])
plt.figure(figsize = (12,7))
sn.heatmap(df_cm, annot=True)
```

## AUC-ROC Curve

```python
fpr, tpr, thresholds = metrics.roc_curve(y_pred, y_true, pos_label=9)
roc_auc = auc(fpr, tpr)

plt.figure()
lw = 2
plt.plot(fpr, tpr, color='red',
         lw=lw, label='ROC curve (area = %0.3f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0,])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

## Accuracy on test and train images

```python
correct = 0
total = 0

with torch.no_grad():  # Not calculating gradients.
    for data in testloader:
        images, labels = data
        images = images.to(device)
        labels = labels.to(device)
        # Calculate outputs by running images through the network.
        outputs = net(images)
        # Label with the highest confidence is what model choose as prediction.
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        test_acc = correct / total
        
print('Accuracy of the network on the 10000 test images: %d %%' % (
    100 * correct / total))
```
